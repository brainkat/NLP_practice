### word frequency

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

count_vectorizer = CountVectorizer(
    ngram_range=(1, 2),   
    max_features=50     
)
corpus_vecs = count_vectorizer.fit_transform(corpus_text)  
print(f"vector size {corpus_vecs.toarray().shape}")     # (sentence, word) 

# vocabulary & frequency

word_counts = corpus_vecs.toarray().sum(axis=0)    # frequency calculation     
vocab = sorted(count_vectorizer.vocabulary_.items(), key=lambda x: x[1])

vocab_df = pd.DataFrame(vocab, columns=['word', 'idx'])
vocab_df['count'] = vocab_df['idx'].apply(lambda i: word_counts[i]) 
vocab_df = vocab_df.drop(columns=['idx']) #drop (unnecessary)
vocab_df = vocab_df.sort_values('count', ascending=False).reset_index(drop=True)

print(vocab_df)
```

### TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
text_vecs = tfidf_vectorizer.fit_transform(corpus_text)

print(text_vecs.toarray())
print(tfidf_vectorizer.get_feature_names_out())
```

### cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

corpus_vecs_arr = text_vecs.toarray()
first_similarities = cosine_similarity(corpus_vecs_arr[0:1], corpus_vecs_arr) #first against all
print(first_similarities)

# similarity between all of them
all_similarities = cosine_similarity(corpus_vecs_arr, corpus_vecs_arr) #compare all between all
print(all_similarities)


# most similar to first one
sorted_sentences = first_similarities.argsort()[0][::-1]

# top 10 most similar 
for sen in sorted_sentences[1:11]:  # 0 (self) ommitted
    print(f"similarity 유사도: {first_similarities[0][sen]:.4f}")
    print(f"sentence: {corpus_text[sen]}")
```

### sentiment analysis method 1 
- (negative, positive word  keyword counting method)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 긍/부정 standard
positive_ref = '괜찮다 좋다 좋아 감사 효과 도움 추천 개선 긍정 최고'
negative_ref = '안되다 안좋다 부작용 싫어 싫다 악화 심하다 나쁨'

def predict_sentiment(input_text):
    texts = [positive_ref, negative_ref, input_text]
    vecs = tfidf_vectorizer.transform(input_text)
    
    # input vs positive cosine similarity 
    pos_score = cosine_similarity(vecs[2], vecs[0])[0][0] #[2] input, [0] positive
    
    # input vs negative cosine similarity 
    neg_score = cosine_similarity(vecs[2], vecs[1])[0][0] #[1] negative
    
    sentiment = 'positive' if pos_score > neg_score else 'negative'
    
    return {
        'text': input_text,
        'positive_score': round(pos_score, 4),
        'negative_score': round(neg_score, 4),
        'sentiment': sentiment
    }

result = predict_sentiment(corpus_text)
result
```
