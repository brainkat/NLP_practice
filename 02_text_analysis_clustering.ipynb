### TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
text_vecs = tfidf_vectorizer.fit_transform(corpus_text)

print(text_vecs.toarray())
print(tfidf_vectorizer.get_feature_names_out())
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    max_iter=5000,
    random_state=0
)

opinions_label = kmeans.fit_predict(text_vecs)

centers = kmeans.cluster_centers_
centroid_idx = centers.argsort()[:, ::-1]    #flips columns of every row 

sentiment_map = {}
pos_keywords = {'좋다', '감사하다', '괜찮다', '도움', '낫다'}
neg_keywords = {'안좋다', '나쁨', '줄이다', '끊다', '부작용'}

for i in range(3):
    top_20_idx = centroid_idx[i, :20]
    top_words = set(feature_names[top_20_idx])
    
    pos_hits = len(top_words.intersection(pos_keywords))
    neg_hits = len(top_words.intersection(neg_keywords))
    
    if pos_hits > neg_hits:
        sentiment_map[i] = "Positive"
    elif neg_hits > pos_hits:
        sentiment_map[i] = "Negative"
    else:
        sentiment_map[i] = "Neutral"

    print(f"Cluster {i} identified as: {sentiment_map[i]}")
    print(f"Keywords: {list(top_words)[:10]}\n")
```
